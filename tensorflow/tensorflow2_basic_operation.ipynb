## Tensorflow v2 Basic Operation
- https://github.com/aymericdamien/TensorFlow-Examples

```
import tensorflow as tf
```

### Tensor variables

```
# Constant
hello = tf.constant('Hello World')
a = tf.constant(2)
b = tf.constant(3)
m1 = tf.constant([[1, 2], [3, 4]])
m2 = tf.constant([[3, 4], [1, 2]])
```

### Tensor Operations

```
# Arithmetic (add / subtract / multiply / divide)
c1 = tf.add(a, b)
c2 = tf.multiply(a, b)
m3 = tf.matmul(m1, m2)
```

```
# Statistics (reduce_mean / reduce_sum)
tf.reduce_mean([a, b, c1])
```

```
# Conversion
m3.numpy()
```
